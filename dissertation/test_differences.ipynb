# Analysis

```python
ground_truth_cols = ['A_Scale_score', 'C_Scale_score', 'E_Scale_score', 'N_Scale_score', 'O_Scale_score']
natural_selection_pred_cols = ['a', 'c', 'e', 'n', 'o']
my_pred_cols = ['A_Scale_score_study_pred', 'C_Scale_score_study_pred',	
                'E_Scale_score_study_pred', 'N_Scale_score_study_pred',	'O_Scale_score_study_pred']
```

```python
my_differences = [abs(final[pred] - final[gt]) for pred, gt in zip(my_pred_cols, ground_truth_cols)]
my_differences
```

```python
natural_differences = [abs(final[pred] - final[gt]) for pred, gt in zip(natural_selection_pred_cols, ground_truth_cols)]
natural_differences
```

```python
from scipy import stats
```

```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np

results = []

# Iterate over each target
for i in range(len(my_differences)):
    # Calculate means
    mean_my_differences = np.mean(my_differences[i])
    mean_natural_differences = np.mean(natural_differences[i])
    
    # Calculate differences
    differences = [a - b for a, b in zip(my_differences[i], natural_differences[i])]
    
    # Perform Shapiro-Wilk test
    shapiro_stat, shapiro_p_value = stats.shapiro(differences)
    
    # Calculate pooled standard deviation
    pooled_std = np.sqrt(np.var(my_differences[i], ddof=1) + np.var(natural_differences[i], ddof=1)) / 2
    
    # Calculate Cohen's d
    cohen_d = (mean_my_differences - mean_natural_differences) / pooled_std
    
    # Choose the appropriate test based on normality
    if shapiro_p_value > 0.05:
        # Paired t-test for normally distributed differences
        t_stat, p_value = stats.ttest_rel(my_differences[i], natural_differences[i])
        test_type = 'Paired t-test'
    else:
        # Wilcoxon signed-rank test for non-normally distributed differences
        t_stat, p_value = stats.wilcoxon(my_differences[i], natural_differences[i])
        test_type = 'Wilcoxon signed-rank test'

    # Append the result for each target
    results.append({
        'Target': ground_truth_cols[i],
        'my_differences_mean': mean_my_differences,
        'natural_differences_mean': mean_natural_differences,
        'Shapiro-Wilk P-value': shapiro_p_value,
        'Test Type': test_type,
        'Test Statistic': t_stat,
        'P-value': p_value,
        'Cohen\'s d': cohen_d
    })

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results)

# Display results
print(results_df)
```
